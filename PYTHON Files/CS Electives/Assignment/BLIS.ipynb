```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("BLIS.csv")
df
```

```python
plt.plot(df["Academic Year"], df["1st Semester"], marker='o', label='1st Semester')
plt.plot(df["Academic Year"], df["2nd Semester"], marker='s', label='2nd Semester')
plt.legend()
plt.xlabel("\nAcademic Year")
plt.ylabel("Number of Enrollees\n")
plt.xticks(rotation=30)
plt.show()
```
